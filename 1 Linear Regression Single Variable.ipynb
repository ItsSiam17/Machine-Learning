```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df### y = mx + b, price = m * area + b
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('prices(US $)')
#plt.plot(df.area, df.price, 'r.-')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) #ie, linear reg is ready to predict the price
```

```python
reg.predict([[5000]]) # value of x
```

```python
reg.coef_ #value of m
```

```python
reg.intercept_ #value for b
```

```python
# y = mx + b
135.78767123 * 3300 + 180616.43835616432 # value of y
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('prices(US $)')
#plt.plot(df.area, df.price, 'r.-')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area,reg.predict(df[['area']]), color = 'blue')
plt.show()
```
